```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import sklearn
from ast import literal_eval
```

```python
books = pd.read_csv('goodreads_data.csv', converters={'Genres': literal_eval})
```

```python
books.head()
```

```python
books = books.drop('Unnamed: 0', axis = 1)
```

```python
books
```

```python
philosophy = []
for j in range(books.shape[0] - 1):
    if 'Philosophy' in books['Genres'].iloc[j]:
        philosophy.append(books['Book'].iloc[j])
```

```python
phil_books = books[books['Book'].isin(philosophy)]
```

```python
phil_books = phil_books.reset_index(drop=True)
phil_books
```

```python
to_drop = []
for j in range(phil_books.shape[0] - 1):
    if 'Fiction' in phil_books['Genres'].iloc[j]:
        to_drop.append(j)
        
phil_books = phil_books.drop(to_drop, axis=0)
```

```python
phil_books = phil_books.reset_index(drop=True)
phil_books.to_csv('Phil_books')
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from scipy.cluster import hierarchy as shc
from sklearn.cluster import AgglomerativeClustering

# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)
```

```python
phil_books['Avg_Rating_use'] = pd.to_numeric(phil_books['Avg_Rating'])
```

```python
phil_books['Num_Ratings_use'] = phil_books['Num_Ratings'].str.replace(",", '')
phil_books['Num_Ratings_use'] = pd.to_numeric(phil_books['Num_Ratings_use'])
```

```python
phil_books
```

```python
scaler = StandardScaler()

scaler.fit(phil_books[['Avg_Rating_use']])
phil_books['Avg_Rating_use'] = scaler.transform(phil_books[['Avg_Rating_use']])

scaler.fit(phil_books[['Num_Ratings_use']])
phil_books['Num_Ratings_use'] = scaler.transform(phil_books[['Num_Ratings_use']])
```

```python
k_rng = range(1,10)
sse = [] 

for k in k_rng:
    km = KMeans(n_clusters = k)
    km.fit(phil_books[['Avg_Rating_use', 'Num_Ratings_use']])
    sse.append(km.inertia_)

sse
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
plt.plot(k_rng, sse);
```

```python
km = KMeans(n_clusters = 6)
prediction = km.fit_predict(phil_books[['Avg_Rating_use', 'Num_Ratings_use']])
phil_books['clusters'] = prediction
```

```python
phil_books
```

```python
phil_books1 = phil_books[phil_books["clusters"] == 0]
phil_books2 = phil_books[phil_books["clusters"] == 1]
phil_books3 = phil_books[phil_books["clusters"] == 2]
phil_books4 = phil_books[phil_books["clusters"] == 3]
phil_books5 = phil_books[phil_books["clusters"] == 4]
phil_books6 = phil_books[phil_books["clusters"] == 5]


fig, ax = plt.subplots(figsize = (10,10))

plt.scatter(phil_books1['Avg_Rating_use'] , phil_books1['Num_Ratings_use'], color = 'green');
plt.scatter(phil_books2['Avg_Rating_use'] , phil_books2['Num_Ratings_use'], color = 'blue');
plt.scatter(phil_books3['Avg_Rating_use'] , phil_books3['Num_Ratings_use'], color = 'red');
plt.scatter(phil_books4['Avg_Rating_use'] , phil_books4['Num_Ratings_use'], color = 'orange');
plt.scatter(phil_books5['Avg_Rating_use'] , phil_books5['Num_Ratings_use'], color = 'black');
plt.scatter(phil_books6['Avg_Rating_use'] , phil_books6['Num_Ratings_use'], color = 'yellow');
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='purple', s=200);
```

```python
phil_books_update = phil_books[phil_books['clusters'].isin([1,2,4,5])]
```

```python
phil_books_update = phil_books_update.reset_index(drop=True)
phil_books_update
```

```python
phil_books_update['Genres_use'] = phil_books_update['Genres'].apply(lambda x: ' '.join(x))
```

```python
phil_books_update['Descriptions_Genres'] = phil_books_update['Genres_use'] + " - " + phil_books_update['Description']
```

```python
phil_books_update['Descriptions_Genres'][0]
```

```python
from sentence_transformers import SentenceTransformer, util
embedder = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
desc_genre_embedding = embedder.encode(phil_books_update['Descriptions_Genres'], convert_to_tensor=False)
```

```python
def Find_similar_books(book, amount):
    cosine_scores = util.cos_sim(desc_genre_embedding[phil_books_update.index[phil_books_update['Book'] == book] ], desc_genre_embedding)
    scores = cosine_scores[0].tolist()
    scores = sorted(range(len(scores)), key=lambda k: scores[k], reverse = True)
    return scores[1:amount]
```

```python
phil_books_update = phil_books_update.drop(columns=['Avg_Rating_use', 'Num_Ratings_use', 'clusters', 'Genres_use', 'Descriptions_Genres' ])
```
